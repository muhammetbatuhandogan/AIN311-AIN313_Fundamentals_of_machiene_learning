## Outlier Detection and Handling

Outliers can significantly impact the performance of distance-based algorithms like kNN because they can disproportionately affect distance calculations. To address this, I have implemented an outlier detection and handling strategy using the Interquartile Range (IQR) method. 

### Approach:
1. **Outlier Detection**: I used the IQR method to identify outliers across all numerical features. Values that fall below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR` are considered outliers.
2. **Outlier Replacement**: Instead of removing outliers, i chose to replace them with the median value of the respective feature. This ensures that the dataset remains robust without the extreme values skewing distance calculations.
3. **Scenarios**: The modified dataset with outliers replaced by the median is added as a new scenario to be tested with the kNN algorithm.


```python
# Function to create boxplots for all numerical features
def plot_boxplots(data, columns):
    plt.figure(figsize=(16, 8))
    
    # Create a subplot for each feature
    for i, column in enumerate(columns, 1):
        plt.subplot(2, 4, i)  # Adjust to fit 2 rows, 4 columns layout (or modify as needed)
        sns.boxplot(x=data[column])
        plt.title(f"Boxplot of {column}")
        plt.xlabel(column)
    
    plt.tight_layout()
    plt.show()

# Generate boxplots for all numerical features
plot_boxplots(df, numerical_columns)
```

```python
# Function to detect and print outliers for multiple columns using IQR
def detect_outliers_iqr_all(data, columns):
    outliers_dict = {}
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
        outliers_dict[column] = outliers
        
        # Print the results
        print(f"Outliers in '{column}':")
        print(f"Number of outliers in '{column}': {len(outliers)}\n")
    
    return outliers_dict

# List of numerical columns to check for outliers
numerical_columns = ['address', 'salary', 'employment_status', 'reside']

# Detect outliers across all numerical features
outliers_all_features = detect_outliers_iqr_all(df, numerical_columns)
```

```python
# Function to replace outliers with the median for multiple columns using IQR
def replace_outliers_with_median_all(data, columns):
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Replace outliers with median
        median_value = data[column].median()
        data[column] = data[column].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)
    
    return data

# List of numerical columns to apply the outlier replacement
numerical_columns = ['customer_since', 'age', 'address', 'salary', 'ed', 'employment_status', 'reside']

#df_replaced_all_outliers = replace_outliers_with_median_all(df.copy(), numerical_columns)
```

## k-Nearest Neighbors (kNN) Implementation

In this section, I implement the k-Nearest Neighbors (kNN) algorithm from scratch. The kNN algorithm is a simple, yet effective, classification technique that identifies the `k` nearest data points from the training set to a given test point and predicts the most common label among them.

### Overview:
1. **Euclidean Distance Function**: Measures the distance between two points in feature space.
2. **Standard kNN Function**: Predicts the label for each test point based on the majority vote among the `k` nearest neighbors.
3. **Weighted kNN Function**: An extension of the standard kNN, where neighbors closer to the test point are given more weight in the voting process.


```python
def euclidean_distance(point1, point2):
    """
    Calculate the Euclidean distance between two points.

    Parameters:
    - point1: First point (array-like)
    - point2: Second point (array-like)

    Returns:
    - distance: Euclidean distance between point1 and point2
    """
    return np.sqrt(np.sum((point1 - point2) ** 2))
```

```python
def knn(X_train, y_train, X_test, k):
    """
    Standard k-Nearest Neighbors (kNN) algorithm implementation.

    Parameters:
    - X_train: Feature matrix of the training set (array-like or DataFrame)
    - y_train: Labels of the training set
    - X_test: Feature matrix of the test set (array-like or DataFrame)
    - k: Number of nearest neighbors to consider

    Returns:
    - y_pred: Predicted labels for the test set
    """
    # inputs are numpy arrays for compatibility
    X_train = np.array(X_train)
    X_test = np.array(X_test)
    y_train = np.array(y_train)

    y_pred = []  # The list for result of predictions.
    
    # Loop for each test point
    for test_point in X_test:
        # Calculating distances
        distances = [euclidean_distance(test_point, train_point) for train_point in X_train]
        
        # Get the indices of the k nearest neighbors
        k_indices = np.argsort(distances)[:k]
        
        # Get the labels of the k nearest neighbors
        k_nearest_labels = [y_train[i] for i in k_indices]
        
        # Majority vote for most common label
        majority_label = max(set(k_nearest_labels), key=k_nearest_labels.count)
        y_pred.append(majority_label)
    
    return np.array(y_pred)
```

### Standard kNN Algorithm
The standard kNN algorithm identifies the `k` nearest neighbors by calculating the Euclidean distance between the test point and all points in the training set. It then assigns the most common label among these neighbors as the predicted label for the test point.

**Steps:**
1. Compute the distance between the test point and each training point.
2. Identify the `k` nearest neighbors.
3. Perform a majority vote to predict the label.


```python
def weighted_knn(X_train, y_train, X_test, k):
    """
    Perform weighted k-Nearest Neighbors (kNN) prediction on the test set.
    
    Weights are inversely proportional to the distance: weight = 1/distance.
    
    Parameters:
    - X_train: Feature matrix of the training set (array-like or DataFrame)
    - y_train: Labels of the training set
    - X_test: Feature matrix of the test set (array-like or DataFrame)
    - k: Number of nearest neighbors to consider

    Returns:
    - y_pred: Predicted labels for the test set
    """
    # Ensure inputs are numpy arrays for compatibility
    X_train = np.array(X_train)
    X_test = np.array(X_test)
    y_train = np.array(y_train)

    y_pred = []  # The list for result of predictions.
    
    for test_point in X_test:
        # Calculating distances.
        distances = [euclidean_distance(test_point, train_point) for train_point in X_train]
        
        # Get the indices of the k nearest neighbors
        k_indices = np.argsort(distances)[:k]
        
        # Get the labels and distances of the k nearest neighbors
        k_nearest_labels = [y_train[i] for i in k_indices]
        k_nearest_distances = [distances[i] for i in k_indices]
        
        # Calculate weights (inversely proportional to distance)
        weights = [1 / distance if distance != 0 else 1e6 for distance in k_nearest_distances]  # Avoid division by zero
        
        # Perform weighted voting
        label_weights = {}
        for label, weight in zip(k_nearest_labels, weights):
            if label in label_weights:
                label_weights[label] += weight
            else:
                label_weights[label] = weight
        
        # Predict the label with the highest weighted sum
        majority_label = max(label_weights, key=label_weights.get)
        y_pred.append(majority_label)
    
    return np.array(y_pred)
```

### Weighted kNN Algorithm
The weighted kNN algorithm gives more importance to neighbors that are closer to the test point by assigning them a higher weight. The weight is calculated as the inverse of the distance:
\[
\text{weight} = \frac{1}{\text{distance}}
\]
**Steps:**
1. Compute the distance between the test point and each training point.
2. Identify the `k` nearest neighbors.
3. Calculate weights inversely proportional to the distance.
4. Perform a weighted vote to predict the label.
5. Neighbors that are closer to the test point have a stronger influence on the prediction.
